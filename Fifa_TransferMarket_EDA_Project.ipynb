## ***4. Data Vizualization, Storytelling & Experimenting with charts : Understand the relationships between variables***

#### Chart - 1

````python
# Chart - 1 visualization code
figure=plt.figure(figsize=(9,3))
df.plot(kind='scatter', x='age', y='potential_rating', alpha=0.2, color='red')
plt.title('Potential Rating vs Age')
plt.xlabel('Age')
plt.ylabel('Potential Rating')
````

##### 1. Why did you pick the specific chart?

Answer Here.

Picked a scatter plot because many players could have the same age and potential so to keep that in mind and also show age vs potential relation.


##### 2. What is/are the insight(s) found from the chart?


Answer Here

The graph shows that the a players potential rating isn't just determined by its age but there are other factors as well that affect the potential.


##### 3. Will the gained insights help creating a positive business impact?
Are there any insights that lead to negative growth? Justify with specific reason.

Answer Here

If a club wants to invest in a player they can look at this chart to see how potential rating depends on age and determine whether they want a younger player or an experienced one.

Negative Impact- Since the graph shows that only age doesnt determine the potential a club might invest in an experienced player but the player might degrade due to the factor.

#### Chart - 2

````python
# Chart - 2 visualization code
avg_potential=df.groupby('age')['potential_rating'].mean()
plt.figure(figsize=(10,5))
avg_potential.plot(kind='line', color='blue')
plt.title('Average Potential Rating by Age')
plt.xlabel('Age')
plt.ylabel('Average Potential Rating')
plt.show()
````

##### 1. Why did you pick the specific chart?

Answer Here.

As age is increasing it is easier to undersand with a line graph as the lnie graph shows progression.

##### 2. What is/are the insight(s) found from the chart?

Answer Here

The chart shows the average potential rating at each age. Although the graph and average potential ratings seem pretty steady there is a very sall spike in potential at around 21 years and 28,29. In the end half of the career that is >30 years, the potential seems to dip very slightly but there is no significant increas or decrease in the given graph


##### 3. Will the gained insights help creating a positive business impact?
Are there any insights that lead to negative growth? Justify with specific reason.

Answer Here

The gained insights from the graph can help the clubs looking to sign players determine their potential or at least guess which will help them make a decision as to who they wanna sign for them.
